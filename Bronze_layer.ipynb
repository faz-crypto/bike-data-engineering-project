```python
df=spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format","csv")\
    .option("cloudFiles.schemaLocation","/Volumes/workspace/bronzebike/bronzevolume/bronzedata/checkpoint")\
    .option("cloudFiles.schemaEvolutionMode","rescue")\
    .load("/Volumes/workspace/rawbike/rawvolume/rawdata/")

    
```

```python
df.writeStream.format("delta")\
    .outputMode("append")\
    .trigger(once=True)\
    .option("checkpointLocation","/Volumes/workspace/bronzebike/bronzevolume/bronzedata/checkpoint")\
    .option("path","/Volumes/workspace/bronzebike/bronzevolume/bronzedata/data")\
    .start()
```
